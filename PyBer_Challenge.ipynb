```python
# Dependencies 
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Files to load
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"
```

```python
# Read the city data file and store it in a pandas DataFrame. 
city_data_df = pd.read_csv(city_data_to_load)
city_data_df.head(10)
```

```python
# Read the ride data file and store it in a pandas DataFrame.
ride_data_df = pd.read_csv(ride_data_to_load)
ride_data_df.head(10)
```

```python
# Combine the data into a single dataset 
pyber_data_df= pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])
pyber_data_df.head()
```

```python
# CHALLENGE 5-- PART 1. Create a PyBer Summary DataFrame

#1. Get the total rides, total drivers, and total fares for each city 
total_fares_by_type = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fares_by_type
```

```python
total_rides_by_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides_by_type
```

```python
total_drivers_by_type = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers_by_type
```

```python
# Get the average fare per ride for each city type.
avg_fare_per_ride = total_fares_by_type / total_rides_by_type
avg_fare_per_ride
```

```python
avg_fare_per_driver = total_fares_by_type / total_drivers_by_type
avg_fare_per_driver
```

```python
####!!!!!!!!! DELETE INDEX TYPE !!!!!!!!!!!!
```

```python
# Adding a list of values with keys to create a new dataframe. 
pyber_summary_df = pd.DataFrame(
        {"Total Rides": total_rides_by_type,
         "Total Drivers": total_drivers_by_type,
         "Total Fares": total_fares_by_type,
         "Average Fare per Ride": avg_fare_per_ride,
         "Average Fare per Driver": avg_fare_per_driver})
pyber_summary_df
```

```python
# Format the columns

pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,}".format)

pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)

pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)

pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)

pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)

pyber_summary_df
```

```python
# PART 2. Create a multiple-line plot for the sum of the fares for each city type.

## Combine the dataset into a single dataset
pyber_df = pd.merge(pyber_data_df, pyber_summary_df, how="left", on=["type"])
pyber_df                    
```

```python
# Reorder the column
new_column_order =["type", "city", "date","fare", "ride_id", "driver_count", "Total Rides", "Total Drivers", "Total Fares", "Average Fare per Ride", "Average Fare per Driver"]

#Assign pyber df the new column order
pyber_df = pyber_df[new_column_order]
pyber_df
```

```python
# Rename columns
pyber_df.rename(columns ={'city':'City', 'date':'Date', 'fare':'Fare', 'ride_id':'Ride Id', 'driver_count':'No. Drivers', 'type':'City Type'}, inplace= True)
pyber_df
```

```python
pyber_copy = pyber_df[["Date", "City Type", "Fare"]].copy()

pyber_copy
```

```python
# Converting the index as datetime

pyber_copy.set_index("Date")
```

```python
t= pyber_copy["Date"] 
pyber_copy.dtypes
pyber_copy['Date'] = pd.to_datetime(pyber_copy['Date'])
print (pyber_copy['Date'].dtype)
```

```python
pyber_copy['Year'] = pyber_copy['Date'].dt.year

pyber_copy['Month'] =pyber_copy['Date'].dt.month

pyber_copy['Day'] = pyber_copy['Date'].dt.day

pyber_copy
```

```python
pyber_copy.info()
```

```python
# Calculate the sum() of fares by the type of city and date using groupby() to create a new DataFrame.
sum_fare_by_date_type = pyber_copy.groupby(["City Type","Date"]).sum()["Fare"]
sum_fare_by_date_type
```

```python
#Reset the index
sum_fare_by_date_type.reset_index()
#Reassigning 
sum_fare_by_date_type = sum_fare_by_date_type.reset_index()
sum_fare_by_date_type
```

```python
#Pivot table 
table = pd.pivot_table(sum_fare_by_date_type, index=["Date"], columns=["City Type"])
table
```

```python
newtable=table.loc['2019-01-01':'2019-04-28']
newtable
```

```python
# Resampling into weekly bins 
weekly_resample = newtable.resample("W").sum()["Fare"]
weekly_resample
```

```python
#Plot using the object-oriented interface method 

plt.style.use('fivethirtyeight')

ax= weekly_resample.plot(figsize=(15,6))
ax.set_xlabel("Month",fontsize=14, fontweight='bold')
ax.set_ylabel("Fare ($USD)",fontsize=14, fontweight='bold')
ax.set_title("Total Fare by City Type",fontsize=20, fontweight='bold')

#Save the figure
plt.savefig("analysis/Fig.challenge.png")
plt.show()
```
