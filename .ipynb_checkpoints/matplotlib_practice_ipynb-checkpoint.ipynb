```python
%matplotlib inline
```

```python
# Import dependencies. 
import matplotlib.pyplot as plt 
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Object-Oriented Interface

# Create the plot with ax.plt() - #1
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() --#2
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() -- quickest method 
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend. 
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0-45.
plt.ylim(0,45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot and add a label for the legend. 
plt.plot(x_axis, y_axis, marker="*", color = "blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0-45.
plt.ylim(0,45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid. 
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Object-oriented approach:
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label="Boston")
# Create labels for the x and y axes.
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
# Set the y limit between 0-45.
ax.set_ylim(0,45)
# Create a title.
ax.set_title("PyBer Fare by Month")
# Add a grid. 
ax.grid()
# Add the legend.
ax.legend()
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot
plt.bar(x_axis, y_axis, color ="green", label="Boston")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title 
plt.title("PyBer Fare by Month")
# Add the legened 
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis)
```

```python
# Opposite axes 
plt.barh(y_axis, x_axis)
```

```python
# Create the plot
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# Skill Drill 
# Create the plot
plt.barh(x_axis, y_axis, color ="purple", label="Boston")
plt.gca().invert_yaxis()
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title 
plt.title("PyBer Fare by Month")
# Add the legened 
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color='k', label='New York')
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
# opposite axes
fix, ax = plt.subplots()
ax.barh(y_axis, x_axis)
```

```python
# Skill Drill 
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
ax.invert_yaxis()
ax.set_xlabel("Date")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare by Month")
ax.legend()
```

```python
# Create a Scatter Plot using MATLAB plt.plot() & plt.scatter()
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(x_axis, y_axis)
```

```python
plt.scatter(x_axis, y_axis, c='red', label='Chicago')
plt.gca().invert_yaxis()
plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.xlim(0,50)
plt.title("PyBer Fare by Month")
plt.legend()
```

```python
# Bubble Chart using MATLAB method 
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger =[]
for data in y_axis:
    y_axis_larger.append(data*3)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Create a Pie Chart using the MATLAB Method 
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
plt.subplots(figsize=(8,8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```
